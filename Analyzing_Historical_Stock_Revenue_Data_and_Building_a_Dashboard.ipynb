```python
import yfinance as yf
import matplotlib.pyplot as plt

# Step 1: Download Tesla stock data
tesla_data = yf.download('TSLA', start='2020-01-01', end='2024-01-01')

# Step 2: Create a function to plot the graph
def make_graph(data):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], label='Tesla Stock Price', color='blue')
    plt.title('Tesla Stock Price Over Time')  # Add title to the graph
    plt.xlabel('Date')  # Label for the x-axis
    plt.ylabel('Price (USD)')  # Label for the y-axis
    plt.legend()  # Show legend
    plt.grid(True)  # Show grid for better readability
    plt.xticks(rotation=45)  # Rotate date labels for clarity
    plt.tight_layout()  # Ensure everything fits without overlapping
    plt.show()

# Step 3: Call the function to plot the graph
make_graph(tesla_data)
```
